```python
import numpy as np
import pandas as pd

# PREPARE DATA

# Copy dataframe
data = pd.read_csv("/kaggle/input/glass/glass.csv")

# Binary classification:
# Class 1 vs Rest (change 1 to any class you want)
data['Type'] = data['Type'].apply(lambda x: 1 if x == 1 else 0)

# Separate features and target
X = data.drop('Type', axis=1).values
y = data['Type'].values

# Feature scaling (important for gradient descent)
X = (X - X.mean(axis=0)) / X.std(axis=0)

# Train-test split (manual, no sklearn)
split_ratio = 0.8
split_index = int(len(X) * split_ratio)

X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]


# INITIALIZE PARAMETERS

num_features = X_train.shape[1]
weights = np.zeros(num_features)
bias = 0.0

learning_rate = 0.01
epochs = 1000

# SIGMOID FUNCTION

def sigmoid(z):
    return 1 / (1 + np.exp(-z))


# TRAINING LOOP (PERCEPTRON VIEW)

for epoch in range(epochs):

    total_loss = 0

    for i in range(len(X_train)):

        # Forward pass
        z = np.dot(weights, X_train[i]) + bias
        y_pred = sigmoid(z)

        # Binary Cross-Entropy Loss
        loss = - (y_train[i] * np.log(y_pred + 1e-9) +
                  (1 - y_train[i]) * np.log(1 - y_pred + 1e-9))
        total_loss += loss

        # Error
        error = y_pred - y_train[i]

        # Weight update
        weights = weights - learning_rate * error * X_train[i]
        bias = bias - learning_rate * error

    # Print loss every 100 epochs
    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {total_loss/len(X_train):.4f}")

# PREDICTION FUNCTION

def predict(X):
    predictions = []
    for x in X:
        z = np.dot(weights, x) + bias
        y_hat = sigmoid(z)
        predictions.append(1 if y_hat >= 0.5 else 0)
    return np.array(predictions)

# 6. EVALUATION

y_pred = predict(X_test)

accuracy = np.mean(y_pred == y_test)
print("\nTest Accuracy:", accuracy)
```
